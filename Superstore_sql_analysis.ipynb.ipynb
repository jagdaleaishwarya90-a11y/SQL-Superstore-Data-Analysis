## Superstore sales data analysis using AQL

## Project Overview

This project analysis a retail superstore dataset using SQL queries.The goal is to SQL understand sales performance, customer behavior and product trends.

```python
import sqlite3
import pandas as pd
```

```python
df=pd.read_csv("superstore.csv")
df.head()
```

## SQL dataset create

```python
conn=sqlite3.connect("superstore.db")
df.to_sql("sales",conn,if_exists="replace",index=False)
```

## Analysis Queries

## 1] Total sales

```python
query = """
SELECT SUM(sales)AS tota_sales
FROM sales
"""
pd.read_sql(query,conn)
```

## 2] Total Profit

```python
query ="""
SELECT SUM(profit)AS total_profit
FROM sales
"""
pd.read_sql_query(query,conn)
```

## 3] Category Wise Sales

```python
query ="""
SELECT category,
sum(sales)AS total_sales
FROM sales
GROUP BY category
ORDER BY total_sales DESC
"""
pd.read_sql_query(query,conn)
```

## 4] Sub-Category Sales

```python
query="""
SELECT"Sub-Category",
SUM(sales)AS total_sales
FROM sales
GROUP BY "Sub-Category"
ORDER BY total_sales DESC
"""
pd.read_sql_query(query,conn)
```

## 5] Region Wise Sales

```python
query="""
SELECT Region,
SUM(Sales)AS total_sales
FROM sales
GROUP BY Region
"""
pd.read_sql_query(query,conn)
```

## 6] Top 10 Products

```python
query="""
SELECT"Product Name",
SUM(Sales)AS total_sales
FROM sales
GROUP BY "Product Name"
ORDER BY total_sales DESC
LIMIT 10
"""
pd.read_sql_query(query,conn)
```

## 7]Top Customers

```python
query="""
SELECT "Customer Name",
SUM(Sales)AS total_spent
FROM sales
GROUP BY "Customer Name"
ORDER BY total_spent DESC
LIMIT 10
"""
pd.read_sql_query(query,conn)
```

## 8] Average Sales per Order

```python
query="""
SELECT AVG(Sales)AS avg_sales
FROM sales
"""
pd.read_sql(query,conn)
```

## 9] Highest Profit Products

```python
query="""
SELECT"Product Name",
SUM(Profit)AS total_profit
FROM sales
GROUP BY "Product Name"
ORDER BY total_profit DESC
LIMIT 10
"""
pd.read_sql_query(query,conn)
```

## 10] Loss Marketing Products

```python
query="""
SELECT"Product Name",
SUM(Profit)AS total_profit
FROM sales
GROUP BY "Product Name"
HAVING total_profit < 0
ORDER BY total_profit
"""
pd.read_sql_query(query,conn)
```

## Insights
1] Technology category generates highest sales
2] Some products generate negative profit
